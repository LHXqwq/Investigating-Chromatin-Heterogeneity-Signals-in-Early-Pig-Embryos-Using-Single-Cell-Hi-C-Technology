```python
#PA 2C 4C
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import SVG, display

plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['svg.fonttype'] = 'none'
plt.rcParams['svg.hashsalt'] = 'hello'

# Update the labels as requested
labels = ['2C', '4C']
low_greater_high = [830, 2475]
low_less_high = [693, 627]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

# Update the colors array to include different colors for each bar
colors = ['#7dace4', 'green', '#ff165d', 'orange']

fig, ax = plt.subplots(figsize=(5, 5))  # 设置图形的尺寸为宽度 5，高度 5

# Create bars with different colors
rects1 = ax.bar(x - width/2, low_greater_high, width, color=colors[:2], label='Cluster1 > Cluster2')
rects2 = ax.bar(x + width/2, low_less_high, width, color=colors[2:], label='Cluster1 < Cluster2')

# 设置所有字体大小为25
ax.set_ylabel('Number of bins', fontsize=15)
ax.set_title('PA: Significant Bins with Boundary Frequency Differences', fontsize=15)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=12)
ax.set_ylim(0, 2800)  # 设置 y 轴坐标轴范围
# Add the legend with the specified labels and colors
custom_lines = [
    plt.Line2D([0], [0], color='#7dace4', lw=4),
    plt.Line2D([0], [0], color='#ff165d', lw=4),
    plt.Line2D([0], [0], color='green', lw=4),
    plt.Line2D([0], [0], color='orange', lw=4)
]

ax.legend(custom_lines, [
    'PA-2C-H1 > PA-2C-H2', 
    'PA-2C-H1 < PA-2C-H2', 
    'PA-4C-H1 > PA-4C-H2', 
    'PA-4C-H1 < PA-4C-H2'
], fontsize=10, loc='upper left')


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom',
                    fontsize=15)


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.savefig('PA_boundary_stat.svg', dpi=300, bbox_inches='tight', transparent=True)
```
