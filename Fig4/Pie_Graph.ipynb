```python
#PA 2C 40k TAD
import matplotlib.pyplot as plt
from IPython.display import SVG, display

plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['svg.fonttype'] = 'none'
plt.rcParams['svg.hashsalt'] = 'hello'

# 数据
sizes = [66, 34, 6]
labels = ['PA-2C-H1', 'PA-2C-H2', 'PA-2C-Undefined']
colors = ['#7dace4', '#ff165d', 'lightgrey']  # 自定义颜色

# 自定义autopct格式化函数
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return f'{val} ({pct:.1f}%)'
    return my_format

# 创建饼图
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(sizes, labels=labels, colors=colors, autopct=autopct_format(sizes), startangle=90)

# 设置标注信息的样式
for text in texts:
    text.set_fontsize(20)
for autotext in autotexts:
    autotext.set_color('black')
    autotext.set_fontsize(15)

# 确保饼图是圆的
ax.axis('equal')

# 添加标题
plt.title('PA:2C', fontsize=20,pad=20)

plt.savefig('PA_2C_pie.svg', dpi=300, bbox_inches='tight', transparent=True)
```
