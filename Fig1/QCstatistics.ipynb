```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import SVG, display

plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['svg.fonttype'] = 'none'
plt.rcParams['svg.hashsalt'] = 'hello'

# 阶段
stages = ["MII/Zygote", "2C", "4C"]
x = np.arange(len(stages))

# 每组的通过数和总数
pass_counts = {
    "PA": [35, 117, 203],
    "IVF": [31, 109, 210]
}
total_counts = {
    "PA": [69, 234, 338],
    "IVF": [48, 190, 281]
}

width = 0.35  # 柱宽

# 创建图形
fig, ax = plt.subplots(figsize=(8, 6))

# 绘制柱状图
bar1 = ax.bar(x - width/2, pass_counts["PA"], width, label='PA', color='skyblue')
bar2 = ax.bar(x + width/2, pass_counts["IVF"], width, label='IVF', color='lightgreen')

# 添加标签（通过数/总数）
for i in range(len(stages)):
    ax.text(x[i] - width/2, pass_counts["PA"][i] + 5,
            f'{pass_counts["PA"][i]}/{total_counts["PA"][i]}', ha='center', fontsize=15)
    ax.text(x[i] + width/2, pass_counts["IVF"][i] + 5,
            f'{pass_counts["IVF"][i]}/{total_counts["IVF"][i]}', ha='center', fontsize=15)

# 设置图像属性
ax.set_ylabel('Pass Count', fontsize=20)
ax.set_title('scHi-C: Embryo Stage Pass Counts', fontsize=20)
ax.set_xticks(x)
ax.set_xticklabels(stages, fontsize=20)
ax.tick_params(axis='y', labelsize=20)
ax.legend(fontsize=20)
ax.set_ylim(0, 225)

plt.tight_layout()
plt.savefig('scHi-C_QC_count.svg', dpi=300, bbox_inches='tight', transparent=True)
```
