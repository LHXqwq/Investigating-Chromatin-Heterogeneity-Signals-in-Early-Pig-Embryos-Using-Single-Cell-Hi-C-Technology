```python
#PA QC
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from IPython.display import SVG, display

# sum.txt
data_column2 = []
data_column3 = []
data_column4 = []
data_column5 = []
with open('PA_QC.txt', 'r') as file:
    next(file)
    for line in file:
        line = line.strip().split('\t')
        if len(line) == 5:
            data_column2.append(int(line[1]))
            data_column3.append(float(line[2]))
            data_column4.append(float(line[3]))
            data_column5.append(float(line[4]))

fig, axs = plt.subplots(1, 4, figsize=(20, 5))
sns.set_theme(style="white")
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['svg.fonttype'] = 'none'
plt.rcParams['svg.hashsalt'] = 'hello'

sns.histplot(data_column2,ax=axs[0],bins=40,kde=True)
axs[0].set_xlabel('Valid contacts')
#axs[0].set_xscale("log")
axs[0].set_ylabel('Counts')
axs[0].axvline(x=30000, linestyle='--', color='red')

sns.histplot(data_column3,ax=axs[1],bins=40,kde=True)
axs[1].set_xlabel('Trans/Total contacts')
#axs[1].set_xscale("log")
axs[1].set_ylabel('Counts')
axs[1].axvline(x=0.25, linestyle='--', color='red')
x_ticks, y_ticks = axs[1].get_xticks(), axs[1].get_yticks()
print(x_ticks, y_ticks)

sns.histplot(data_column4,ax=axs[2],bins=40,kde=True)
axs[2].set_xlabel('Coverage')
axs[2].set_ylabel('Counts')
#axs[2].set_xscale("log")
axs[2].axvline(x=0.8, linestyle='--', color='red')

sns.histplot(data_column5,ax=axs[3],bins=40,kde=True)
axs[3].set_xlabel('GiniQC')
axs[3].set_ylabel('Counts')
#axs[3].set_xscale("log")
axs[3].axvline(x=0.55, linestyle='--', color='red')

plt.savefig('PA_QC.svg', dpi=300, bbox_inches='tight', transparent=True)
display(SVG(filename='PA_QC.svg'))
```

```python
# PA+IVF QC
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from IPython.display import SVG, display

sns.set_theme(style="white")
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['svg.fonttype'] = 'none'
plt.rcParams['svg.hashsalt'] = 'hello'

# 
data_column2 = []
data_column3 = []
data_column4 = []
data_column5 = []
with open('PA_IVF_QC.txt', 'r') as file:
    next(file)  # 
    for line in file:
        line = line.strip().split('\t')
        if len(line) == 5:
            data_column2.append(int(line[1]))
            data_column3.append(float(line[2]))
            data_column4.append(float(line[3]))
            data_column5.append(float(line[4]))

# 22figsize
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
sns.set_theme(style="white")

# 
plt.subplots_adjust(hspace=0.3, wspace=0.3)

#  (0,0)
sns.histplot(data_column2, ax=axs[0,0], bins=40, kde=True)
axs[0,0].set_xlabel('Valid contacts')
axs[0,0].set_ylabel('Counts')
axs[0,0].axvline(x=30000, linestyle='--', color='red')

#  (0,1)
sns.histplot(data_column3, ax=axs[0,1], bins=40, kde=True)
axs[0,1].set_xlabel('Trans / Total contacts')
axs[0,1].set_ylabel('Counts')
axs[0,1].axvline(x=0.25, linestyle='--', color='red')

#  (1,0)
sns.histplot(data_column4, ax=axs[1,0], bins=40, kde=True)
axs[1,0].set_xlabel('Coverage')
axs[1,0].set_ylabel('Counts')
axs[1,0].axvline(x=0.8, linestyle='--', color='red')

#  (1,1)
sns.histplot(data_column5, ax=axs[1,1], bins=40, kde=True)
axs[1,1].set_xlabel('GiniQC')
axs[1,1].set_ylabel('Counts')
axs[1,1].axvline(x=0.55, linestyle='--', color='red')

# 
plt.tight_layout()

plt.savefig('IVF_PA_QC.svg', dpi=300, bbox_inches='tight', transparent=True)
```
