```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import SVG, display

plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['svg.fonttype'] = 'none'
plt.rcParams['svg.hashsalt'] = 'hello'

# Load the data
data = pd.read_table('PA_Cell_Cycle.txt', sep='\t')

# Set up the figure with two subplots
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Set white background and remove grid
sns.set(style="white")


# Custom palette for Cell Cycle
cell_cycle_palette = {
    "G1": "#c2a5cf",
    "Early-S": "#99d8c9",
    "Late_S-G2": "#41ae76",
    "M": "#ea6e34"
}

# Scatter plot for Cell Cycle
g1 = sns.scatterplot(
    x="2-12M%",  # Specify the x-axis data
    y="<2m%",  # Specify the y-axis data
    data=data,  # Specify the data set to use
    hue="Cell_Cycle", palette=cell_cycle_palette,
    ax=ax1, s=20
)

g1.set_title('PA Cell Cycle', fontsize=20, color='black')
g1.set_xlabel('% Mitotic contacts', fontsize=20, color='black')
g1.set_ylabel('% Short-range contacts', fontsize=20, color='black')
g1.set_yticks([0.5, 0.7, 0.9])
g1.set_yticklabels(['50', '70', '90'])  # Replace tick labels with 'a', 'b', 'c'
g1.set_xticks([0.1, 0.3])
g1.set_xticklabels(['10','30'])  # Replace tick labels with 'a', 'b', 'c'
g1.tick_params(axis='both', labelsize=15, colors='black')  # Set font size for both x and y ticks
# Set the legend manually to ensure order
handles, labels = ax1.get_legend_handles_labels()
order = ["G1", "Early-S", "Late_S-G2", "M"]
sorted_handles_labels = sorted(zip(handles, labels), key=lambda x: order.index(x[1]))
handles, labels = zip(*sorted_handles_labels)
g1.legend(handles=handles, labels=labels, title='Cell Cycle', title_fontsize=15, fontsize=15, loc='upper right')

# Custom palette for Cell Type
cell_type_palette = {
    "PA-MII": "#ffe119",
    "PA-2C": "#3cb44b",
    "PA-4C": "#e6194b"
}

# Scatter plot for Cell Type
g2 = sns.scatterplot(
    x="2-12M%",  # Specify the x-axis data
    y="<2m%",  # Specify the y-axis data
    data=data,  # Specify the data set to use
    hue="Cell_Type", palette=cell_type_palette,
    ax=ax2, s=20
)

g2.set_title('PA Cell Type', fontsize=20, color='black')
g2.set_xlabel('% Mitotic contacts', fontsize=20, color='black')
g2.set_ylabel('% Short-range contacts', fontsize=20, color='black')
g2.set_yticks([0.5, 0.7, 0.9])
g2.set_yticklabels(['50', '70', '90'])  # Replace tick labels with 'a', 'b', 'c'
g2.set_xticks([0.1, 0.3])
g2.set_xticklabels(['10','30'])  # Replace tick labels with 'a', 'b', 'c'
g2.tick_params(axis='both', labelsize=15, colors='black')  # Set font size for both x and y ticks
# Set the legend manually to ensure order
handles, labels = ax2.get_legend_handles_labels()
order = ["PA-MII", "PA-2C", "PA-4C"]
sorted_handles_labels = sorted(zip(handles, labels), key=lambda x: order.index(x[1]))
handles, labels = zip(*sorted_handles_labels)
g2.legend(handles=handles, labels=labels, title='Cell Type', title_fontsize=15, fontsize=15, loc='upper right')

# Adjust layout
plt.tight_layout()

#  PDF
plt.savefig('PA_Cell_Cycle.svg', dpi=300, bbox_inches='tight', transparent=True)
#display(SVG(filename='PA_Cell_Cycle.svg'))
```
