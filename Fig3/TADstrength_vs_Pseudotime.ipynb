```python
#IVF 去掉 极端百分位（如 2.5% 和 97.5% 之外的值）：
import glob
import numpy as np
import os

output_file = "/home/lihaoxing/PA_IVF_532_25w_TAD/aggregateTADs/IVF_strength/IVF_strength_trimmed_mean.txt"

with open(output_file, "w") as out:
    out.write("Cell\tTrimmedMeanValue\n")
    for filename in glob.glob("/home/lihaoxing/PA_IVF_532_25w_TAD/aggregateTADs/IVF_strength/*.strength.txt"):
        values = []
        with open(filename) as f:
            for line in f:
                parts = line.strip().split("\t")
                try:
                    val = float(parts[4])
                    if np.isfinite(val) and val >= 0:  # 过滤掉 inf, -inf, 以及 < 0 的值
                        values.append(val)
                except ValueError:
                    continue
        if values:
            values = np.array(values)
            lower_bound, upper_bound = np.percentile(values, [2.5, 97.5])  # 计算 1% 和 99% 分位数
            filtered_values = values[(values >= lower_bound) & (values <= upper_bound)]  # 仅保留范围内的值
            trimmed_mean_value = np.mean(filtered_values) if len(filtered_values) > 0 else "NA"
        else:
            trimmed_mean_value = "NA"
        
        cell_name = os.path.basename(filename)  # 只保留文件名
        out.write(f"{cell_name}\t{trimmed_mean_value}\n")

print(f"截尾均值计算完成，结果保存在 {output_file}")
```

```python
#IVF
import pandas as pd
import numpy as np
from scipy.stats import pearsonr, spearmanr

# 读取数据
tad_strength = pd.read_csv("/home/lihaoxing/PA_IVF_532_25w_TAD/aggregateTADs/IVF_strength/individual_files/IVF_merged_tad_strength_matrix_new.txt", sep="\t", index_col=0)
time_data = pd.read_csv("/home/lihaoxing/PA_IVF_532_25w_TAD/aggregateTADs/TADstrength_vs_Time/ivf_time.txt", sep="\t")

# 构建时间字典
time_dict = dict(zip(time_data['id'], time_data['time']))

# 存储相关性结果
correlation_results = []
correlation_results_not_significant = []

for window_id, row in tad_strength.iterrows():
    # 筛选出有数值的细胞
    valid_cells = row.dropna().index

    # 如果该窗口有有效的细胞ID
    if len(valid_cells) > 0:
        # 找到有效细胞的时间值
        times = pd.Series(time_dict).loc[valid_cells].dropna()  # 确保时间数据有效
        
        # 提取对应的窗口强度值
        values = row[valid_cells].loc[times.index].astype(float)  # 保证对应细胞的强度值和时间值
        
        # 过滤掉值为0的细胞
        mask = values != 0
        values = values[mask]
        times = times[mask]
        
        # 确保数据中没有 NaN 或 inf
        values = values.replace([np.inf, -np.inf], np.nan).dropna()
        times = times.replace([np.inf, -np.inf], np.nan).dropna()

        # 重新检查数据是否有效
        if len(values) > 1 and len(values) == len(times):
            num_valid_cells = len(values)  # 记录有效细胞数量

            # 计算皮尔逊和斯皮尔曼相关性
            pearson_corr, pearson_p = pearsonr(values, times)
            spearman_corr, spearman_p = spearmanr(values, times)

            # 仅存储 p 值显著的结果（p < 0.05）
            if pearson_p < 0.05:
                correlation_results.append([window_id, num_valid_cells, pearson_corr, pearson_p, spearman_corr, spearman_p])
            else:
                correlation_results_not_significant.append([window_id, num_valid_cells, pearson_corr, pearson_p, spearman_corr, spearman_p])

# 转换为 DataFrame 并按 Pearson 相关性排序
correlation_df = pd.DataFrame(correlation_results, columns=['Genes', 'Valid_Cell_Count', 'Pearson_corr', 'Pearson_p', 'Spearman_corr', 'Spearman_p'])
correlation_df = correlation_df.sort_values(by='Pearson_corr', ascending=False)

# 输出显著性结果
print(correlation_df)
correlation_df.to_csv("/home/lihaoxing/PA_IVF_532_25w_TAD/aggregateTADs/TADstrength_vs_Time/IVF_correlation_results_significant_genes.txt", sep="\t", index=False)

# 处理非显著性结果
not_correlation_df = pd.DataFrame(correlation_results_not_significant, columns=['Genes', 'Valid_Cell_Count', 'Pearson_corr', 'Pearson_p', 'Spearman_corr', 'Spearman_p'])
not_correlation_df = not_correlation_df.sort_values(by='Pearson_corr', ascending=False)
not_correlation_df.to_csv("/home/lihaoxing/PA_IVF_532_25w_TAD/aggregateTADs/TADstrength_vs_Time/IVF_correlation_results_not_significant_genes.txt", sep="\t", index=False)

print("计算完成，结果已保存。")
```

```python
#IVF 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min

# 读取数据
tad_strength = pd.read_csv(
    "/home/lihaoxing/PA_IVF_532_25w_TAD/aggregateTADs/IVF_strength/individual_files/IVF_merged_tad_strength_matrix_new.txt",
    sep="\t", index_col=0
)

# 读取时间对应关系文件
time_df = pd.read_csv('/home/lihaoxing/PA_IVF_532_25w_TAD/aggregateTADs/IVF_strength/individual_files/ivf_time_sort.txt', sep='\t', header=None, names=['sample', 'time'])
sample_to_time = dict(zip(time_df['sample'], time_df['time']))

# 找到时间在12到14之间的样本
samples_to_remove = [sample for sample, time in sample_to_time.items() if 12.2 <= time <= 13.8]

# 丢掉时间在12到14之间的细胞
tad_strength = tad_strength.drop(columns=samples_to_remove, errors='ignore')

# ✅ 过滤掉 >0 的值少于 15 个的行
tad_strength = tad_strength[tad_strength.gt(0).sum(axis=1) >= 15]

# 将数据转换为 numpy 矩阵用于 KMeans
data = tad_strength.values

# 手肘法确定最佳聚类数
inertia = []
K_range = range(2, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto').fit(data)
    inertia.append(kmeans.inertia_)

# 绘制手肘图
plt.figure(figsize=(6, 4))
plt.plot(K_range, inertia, 'o-')
plt.xlabel('Cluster number (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.tight_layout()
plt.show()

from sklearn.metrics import silhouette_score

silhouette_scores = []
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto').fit(data)
    score = silhouette_score(data, kmeans.labels_)
    silhouette_scores.append(score)

plt.figure(figsize=(6, 4))
plt.plot(K_range, silhouette_scores, 'o-')
plt.xlabel('Cluster number (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Analysis for Optimal k')
plt.grid(True)
plt.tight_layout()
plt.show()

#解释： 轮廓系数越高，聚类效果越好。它是衡量聚类紧密度和分离度的综合指标。
```

```python
#IVF 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import linkage, leaves_list
from scipy.signal import savgol_filter

# 读取数据
tad_strength = pd.read_csv(
    "/home/lihaoxing/PA_IVF_532_25w_TAD/aggregateTADs/IVF_strength/individual_files/IVF_merged_tad_strength_matrix_new.txt",
    sep="\t", index_col=0
)

# 读取时间对应关系文件
time_df = pd.read_csv('/home/lihaoxing/PA_IVF_532_25w_TAD/aggregateTADs/IVF_strength/individual_files/ivf_time_sort.txt', sep='\t', header=None, names=['sample', 'time'])
sample_to_time = dict(zip(time_df['sample'], time_df['time']))

# 找到时间在12到14之间的样本
samples_to_remove = [sample for sample, time in sample_to_time.items() if 12.2 <= time <= 13.8]

# 丢掉时间在12到14之间的细胞
tad_strength = tad_strength.drop(columns=samples_to_remove, errors='ignore')

# ✅ 过滤掉 >0 的值少于 15 个的行
tad_strength = tad_strength[tad_strength.gt(0).sum(axis=1) >= 15]

# 聚类
optimal_k = 3
data = tad_strength.copy()
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init='auto')
cluster_labels = kmeans.fit_predict(data)

# 添加聚类信息
tad_strength['cluster'] = cluster_labels

# 🔥 每个 cluster 内部层级聚类排序
clustered_data = []
for i in range(optimal_k):
    cluster_block = tad_strength[tad_strength['cluster'] == i].drop(columns='cluster')
    # 层次聚类，获取行排序
    if len(cluster_block) > 1:
        linkage_matrix = linkage(cluster_block, method='ward')
        sorted_idx = leaves_list(linkage_matrix)
        cluster_block = cluster_block.iloc[sorted_idx]
    clustered_data.append(cluster_block)

# 合并排序后的数据
data_sorted = pd.concat(clustered_data)
tad_strength_sorted = data_sorted.copy()

# 🔥 绘制热图
plt.figure(figsize=(10, 8))
sns.heatmap(tad_strength_sorted, cmap='vlag', yticklabels=False, xticklabels=True)
plt.title(f'TAD Strength Heatmap with Clustering (k={optimal_k})')
plt.xlabel('Cells (ordered by pseudotime)')
plt.ylabel('Genes (hierarchical-clustered within clusters)')
plt.tight_layout()
plt.show()
```

```python
#IVF
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from scipy.signal import savgol_filter
from IPython.display import SVG, display

plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['svg.fonttype'] = 'none'
plt.rcParams['svg.hashsalt'] = 'hello'

# ✅ 读取样本时间对应关系
time_df = pd.read_csv('/home/lihaoxing/PA_IVF_532_25w_TAD/aggregateTADs/IVF_strength/individual_files/ivf_time_sort.txt', sep='\t', header=None, names=['sample', 'time'])
sample_to_time = dict(zip(time_df['sample'], time_df['time']))

# 📈 每个cluster的平均表达曲线 + 平滑，并计算与时序的相关性
plt.figure(figsize=(8, 6))

correlations = {}  # 存储相关性结果

for i in range(optimal_k):
    cluster_data = tad_strength[tad_strength['cluster'] == i].drop(columns='cluster')
    cluster_mean = cluster_data.mean(axis=0)  # 该 cluster 中每个样本的平均值（index 是样本名）

    # 获取样本时间（只保留在 ivf_time_sort.txt 中有时间信息的）
    sample_names = cluster_mean.index.tolist()
    time_values = []
    mean_values = []
    for sample in sample_names:
        if sample in sample_to_time:
            time_values.append(sample_to_time[sample])
            mean_values.append(cluster_mean[sample])


    # 转为 numpy 数组并按时间排序
    time_values = np.array(time_values)
    mean_values = np.array(mean_values)
    sorted_idx = np.argsort(time_values)
    time_sorted = time_values[sorted_idx]
    mean_sorted = mean_values[sorted_idx]

    # 平滑
    window = min(200, len(mean_sorted) if len(mean_sorted) % 2 == 1 else len(mean_sorted) - 1)
    smoothed = savgol_filter(mean_sorted, window_length=window, polyorder=2)

    # 计算 Pearson 相关性
    r, p = pearsonr(time_sorted, smoothed)
    correlations[f'Cluster {i}'] = (r, p)

    plt.plot(time_sorted, smoothed, label=f'Cluster {i} (r={r:.2f}, p={p:.2e})')

# 图形设置
plt.xlabel('Pseudotime',fontsize=20)
plt.ylabel('Average TAD Strength',fontsize=20)
# 设置刻度字体大小
plt.tick_params(axis='both', labelsize=15)
#plt.title('Smoothed Cluster Mean Curves by Developmental Stage')

# 避免重复 label 的小技巧
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(),fontsize=15)

plt.tight_layout()

# 保存图像为 PDF
plt.savefig('IVF_TADstrength_vs_Pseudotime.svg', dpi=300, bbox_inches='tight', transparent=True)

# 输出相关性结果
print("\n📈 每个 Cluster 曲线与时序的 Pearson 相关性:")
for k, (r, p) in correlations.items():
    print(f"{k}: r = {r:.3f}, p = {p:.2e}")

# 统计每一类的基因数量和占比，并保存基因列表
cluster_counts = tad_strength['cluster'].value_counts().sort_index()
total_genes = len(tad_strength)

print("\n📊 每个 Cluster 的基因数量与占比：")
for i in range(optimal_k):
    count = cluster_counts.get(i, 0)
    percentage = count / total_genes * 100
    print(f"Cluster {i}: {count} genes ({percentage:.2f}%)")

    # 写入文件，每类一个文件
    gene_ids = tad_strength[tad_strength['cluster'] == i].index
    gene_ids.to_series().to_csv(f'IVF_cluster_{i}_genes.txt', index=False, header=False)
```

```python
#PA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import linkage, leaves_list
from scipy.signal import savgol_filter

# 读取数据
tad_strength = pd.read_csv(
    "/home/lihaoxing/PA_IVF_532_25w_TAD/aggregateTADs/PA_strength/individual_files/PA_merged_tad_strength_matrix_new.txt",
    sep="\t", index_col=0
)

# 读取时间对应关系文件
time_df = pd.read_csv('/home/lihaoxing/PA_IVF_532_25w_TAD/aggregateTADs/PA_strength/individual_files/pa_time_sort.txt', sep='\t', header=None, names=['sample', 'time'])
sample_to_time = dict(zip(time_df['sample'], time_df['time']))

# 找到时间在12到14之间的样本
samples_to_remove = [sample for sample, time in sample_to_time.items() if 19 <= time <= 20]

# 丢掉时间在12到14之间的细胞
tad_strength = tad_strength.drop(columns=samples_to_remove, errors='ignore')

# ✅ 过滤掉 >0 的值少于 15 个的行
tad_strength = tad_strength[tad_strength.gt(0).sum(axis=1) >= 10]

# 聚类
optimal_k = 3
data = tad_strength.copy()
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init='auto')
cluster_labels = kmeans.fit_predict(data)

# 添加聚类信息
tad_strength['cluster'] = cluster_labels

# 🔥 每个 cluster 内部层级聚类排序
clustered_data = []
for i in range(optimal_k):
    cluster_block = tad_strength[tad_strength['cluster'] == i].drop(columns='cluster')
    # 层次聚类，获取行排序
    if len(cluster_block) > 1:
        linkage_matrix = linkage(cluster_block, method='ward')
        sorted_idx = leaves_list(linkage_matrix)
        cluster_block = cluster_block.iloc[sorted_idx]
    clustered_data.append(cluster_block)

# 合并排序后的数据
data_sorted = pd.concat(clustered_data)
tad_strength_sorted = data_sorted.copy()

# 🔥 绘制热图
plt.figure(figsize=(10, 8))
sns.heatmap(tad_strength_sorted, cmap='vlag', yticklabels=False, xticklabels=True)
plt.title(f'TAD Strength Heatmap with Clustering (k={optimal_k})')
plt.xlabel('Cells (ordered by pseudotime)')
plt.ylabel('Genes (hierarchical-clustered within clusters)')
plt.tight_layout()
plt.show()
```

```python
#PA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from scipy.signal import savgol_filter
from IPython.display import SVG, display

plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['svg.fonttype'] = 'none'
plt.rcParams['svg.hashsalt'] = 'hello'

# ✅ 读取样本时间对应关系
time_df = pd.read_csv('/home/lihaoxing/PA_IVF_532_25w_TAD/aggregateTADs/PA_strength/individual_files/pa_time_sort.txt', sep='\t', header=None, names=['sample', 'time'])
sample_to_time = dict(zip(time_df['sample'], time_df['time']))

# 📈 每个cluster的平均表达曲线 + 平滑，并计算与时序的相关性
plt.figure(figsize=(8, 6))

correlations = {}  # 存储相关性结果

for i in range(optimal_k):
    cluster_data = tad_strength[tad_strength['cluster'] == i].drop(columns='cluster')
    cluster_mean = cluster_data.mean(axis=0)  # 该 cluster 中每个样本的平均值（index 是样本名）

    # 获取样本时间（只保留在 ivf_time_sort.txt 中有时间信息的）
    sample_names = cluster_mean.index.tolist()
    time_values = []
    mean_values = []
    for sample in sample_names:
        if sample in sample_to_time:
            time_values.append(sample_to_time[sample])
            mean_values.append(cluster_mean[sample])


    # 转为 numpy 数组并按时间排序
    time_values = np.array(time_values)
    mean_values = np.array(mean_values)
    sorted_idx = np.argsort(time_values)
    time_sorted = time_values[sorted_idx]
    mean_sorted = mean_values[sorted_idx]

    # 平滑
    window = min(200, len(mean_sorted) if len(mean_sorted) % 2 == 1 else len(mean_sorted) - 1)
    smoothed = savgol_filter(mean_sorted, window_length=window, polyorder=2)

    # 计算 Pearson 相关性
    r, p = pearsonr(time_sorted, smoothed)
    correlations[f'Cluster {i}'] = (r, p)

    plt.plot(time_sorted, smoothed, label=f'Cluster {i} (r={r:.2f}, p={p:.2e})')

# 图形设置
plt.xlabel('Pseudotime',fontsize=20)
plt.ylabel('Average TAD Strength',fontsize=20)
# 设置刻度字体大小
plt.tick_params(axis='both', labelsize=15)
#plt.title('Smoothed Cluster Mean Curves by Developmental Stage')

# 避免重复 label 的小技巧
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(),fontsize=15)

plt.tight_layout()

# 保存图像为 PDF
plt.savefig('PA_TADstrength_vs_Pseudotime.svg', dpi=300, bbox_inches='tight', transparent=True)

# 输出相关性结果
print("\n📈 每个 Cluster 曲线与时序的 Pearson 相关性:")
for k, (r, p) in correlations.items():
    print(f"{k}: r = {r:.3f}, p = {p:.2e}")

# 统计每一类的基因数量和占比，并保存基因列表
cluster_counts = tad_strength['cluster'].value_counts().sort_index()
total_genes = len(tad_strength)

print("\n📊 每个 Cluster 的基因数量与占比：")
for i in range(optimal_k):
    count = cluster_counts.get(i, 0)
    percentage = count / total_genes * 100
    print(f"Cluster {i}: {count} genes ({percentage:.2f}%)")

    # 写入文件，每类一个文件
    gene_ids = tad_strength[tad_strength['cluster'] == i].index
    gene_ids.to_series().to_csv(f'PA_cluster_{i}_genes.txt', index=False, header=False)
```
