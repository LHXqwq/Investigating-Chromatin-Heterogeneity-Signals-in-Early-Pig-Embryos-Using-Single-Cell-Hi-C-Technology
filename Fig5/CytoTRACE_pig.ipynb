```python
#pre-implant_mouse_embryo
#scRNAseq
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statannotations.Annotator import Annotator
from scipy import stats

# 读取数据
data = pd.read_csv('pre-implant_mouse_embryo_RPKMCytoTRACE_plot_table.txt', sep='\t')

# 设置绘图风格
sns.set(style="whitegrid")
plt.rcParams['font.sans-serif'] = ['DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

trajectory_1 = [
    "early2cell", "mid2cell","late2cell", "4cell","8cell","16cell",   
    "earlyblast","midblast","lateblast"
]

# 合并两个轨迹顺序，轨迹1在前，轨迹2在后
ordered_phenotypes = trajectory_1

# 调整图形的宽度
fig, ax = plt.subplots(figsize=(15, 10))  # 增加宽度

# 创建小提琴图并应用颜色映射和排序
sns.violinplot(x='Phenotype', y='CytoTRACE', data=data,  order=ordered_phenotypes,
               ax=ax, inner="quart", cut=0, scale="width")
sns.stripplot(x='Phenotype', y='CytoTRACE', data=data, order=ordered_phenotypes, 
              color='black', size=5, jitter=True)  # 添加jitter

# 配置统计检验
pairs = [("early2cell", "mid2cell"),("mid2cell", "late2cell"),("late2cell", "4cell"),("4cell", "8cell"),("8cell", "16cell"),
         ("16cell", "earlyblast"),("earlyblast", "midblast"),("midblast", "lateblast")]
annotator = Annotator(ax, pairs, data=data, x="Phenotype", y="CytoTRACE", order=ordered_phenotypes)

# 选择统计检验方法
chosen_test = 'Mann-Whitney'  # 你可以动态设置
annotator.configure(test=chosen_test, text_format='star', line_height=0.01, 
                    line_width=1, loc='outside', fontsize=20)
annotator.apply_and_annotate()

# 旋转 x 轴标签
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylabel('CytoTRACE', fontsize=20)
plt.xlabel('')
plt.tight_layout()  # 调整布局以避免标签被剪裁

# 保存图片为 PDF
# plt.savefig('project_contacts.pdf', format='pdf')

# 显示图形
plt.show()
```

```python
#scRNAseq
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statannotations.Annotator import Annotator
from scipy import stats

# 读取数据
data = pd.read_csv('PA_IVF_198_CytoTRACE_results.txt', sep='\t')

# 设置绘图风格
sns.set(style="whitegrid")
plt.rcParams['font.sans-serif'] = ['DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# 定义配色
color_mapping = {
    "PA-MII": "#ffe119",
    "PA-2C": "#3cb44b",
    "PA-4C": "#e6194b",
    "IVF-Zygote": "#ffe119",
    "IVF-2C": "#3cb44b",
    "IVF-4C": "#e6194b"
}

# 假设有两条发育轨迹的排序信息
# 轨迹1的细胞类型顺序
trajectory_1 = [
    "PA-MII", "PA-2C", "PA-4C"
]

# 轨迹2的细胞类型顺序
trajectory_2 = [
     "IVF-Zygote","IVF-2C", "IVF-4C"
]

# 合并两个轨迹顺序，轨迹1在前，轨迹2在后
ordered_phenotypes = trajectory_1 + trajectory_2

# 调整图形的宽度
fig, ax = plt.subplots(figsize=(15, 6))  # 增加宽度

# 创建小提琴图并应用颜色映射和排序
sns.violinplot(x='Phenotype', y='CytoTRACE', data=data, 
               palette=color_mapping, order=ordered_phenotypes,
               ax=ax, inner="quart", cut=0, scale="width")
sns.stripplot(x='Phenotype', y='CytoTRACE', data=data, order=ordered_phenotypes, color='black', size=5)

# 对数据进行检验
pairs = [("PA-MII", "PA-2C"),("PA-2C", "PA-4C"),("IVF-Zygote", "IVF-2C"),("IVF-2C", "IVF-4C")]
annotator = Annotator(ax, pairs, data=data, x="Phenotype", y="CytoTRACE", order=ordered_phenotypes)

chosen_test='t-test_welch'
if chosen_test == 't-test_ind':
    annotator.configure(test='t-test_ind', text_format='star', line_height=0.01, line_width=1, loc='outside',fontsize=20)
    print("选择了 t-检验 (独立样本 t 检验)。")
elif chosen_test == 't-test_welch':
    annotator.configure(test='t-test_welch', text_format='star', line_height=0.01, line_width=1, loc='outside',fontsize=20)
    print("选择了 Welch's t-检验。")
else:
    annotator.configure(test='Mann-Whitney', text_format='star', line_height=0.01, line_width=1, loc='outside',fontsize=20)
    print("选择了 Mann-Whitney U 检验。")

annotator.apply_and_annotate()

# 旋转 x 轴标签
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylabel('CytoTRACE',fontsize=20)
plt.xlabel('')
plt.tight_layout()  # 调整布局以避免标签被剪裁

# 保存图片为 PDF
# plt.savefig('project_contacts.pdf', format='pdf')

# 显示图形
plt.show()
```

```python
#dip
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statannotations.Annotator import Annotator

# 读取数据
data = pd.read_csv('PA_IVF_198_CytoTRACE_results_cluster2_dip.txt', sep='\t')

# 设置绘图风格
sns.set(style="whitegrid")
plt.rcParams['font.sans-serif'] = ['DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# 定义配色（不改变原始数据标签）
color_mapping = {
    "PA-2C-A1": "#7dace4",
    "PA-2C-A2": "#ff165d",
    "PA-4C-B1": "green",
    "PA-4C-B2": "orange",
    "IVF-2C-C1": "#A73BDD",
    "IVF-2C-C2": "#E7A31B",
    "IVF-4C-D1": "#2292A4",
    "IVF-4C-D2": "#E22720",
}

# 替换横坐标标签的映射
x_label_mapping = {
    "PA-2C-A1": "PA-2C-R1",
    "PA-2C-A2": "PA-2C-R2",
    "PA-4C-B1": "PA-4C-R1",
    "PA-4C-B2": "PA-4C-R2",
    "IVF-2C-C1": "IVF-2C-R1",
    "IVF-2C-C2": "IVF-2C-R2",
    "IVF-4C-D1": "IVF-4C-R1",
    "IVF-4C-D2": "IVF-4C-R2",
}

# 假设有两条发育轨迹的排序信息
# 轨迹1的细胞类型顺序
trajectory_1 = [
    "PA-2C-A1", "PA-2C-A2", "PA-4C-B1", "PA-4C-B2"
]

# 轨迹2的细胞类型顺序
trajectory_2 = [
    "IVF-2C-C1", "IVF-2C-C2", "IVF-4C-D1", "IVF-4C-D2"
]

# 合并两个轨迹顺序，轨迹1在前，轨迹2在后
ordered_phenotypes = trajectory_1 + trajectory_2

# 调整图形的宽度
fig, ax = plt.subplots(figsize=(15, 6))  # 增加宽度

# 创建小提琴图并应用颜色映射和排序
sns.violinplot(x='Phenotype2', y='CytoTRACE', data=data, 
               palette=color_mapping, order=ordered_phenotypes,
               ax=ax, inner="quart", cut=0, scale="width")
sns.stripplot(x='Phenotype2', y='CytoTRACE', data=data, order=ordered_phenotypes, color='black', size=5)

# 对数据进行检验
pairs = [("PA-2C-A1", "PA-2C-A2"), ("PA-4C-B1", "PA-4C-B2"), ("IVF-2C-C1", "IVF-2C-C2"), ("IVF-4C-D1", "IVF-4C-D2")]
annotator = Annotator(ax, pairs, data=data, x="Phenotype2", y="CytoTRACE", order=ordered_phenotypes)

#Mann-Whitney

# t-test_ind, t-test_welch, t-test_paired, Mann-Whitney, Mann-Whitney-gt, Mann-Whitney-ls, Levene, Wilcoxon, Kruskal, Brunner-Munzel.
annotator.configure(test='Mann-Whitney', text_format='star', line_height=0.01, line_width=1, loc='outside', fontsize=20)
annotator.apply_and_annotate()

# 替换x轴标签为新名称
new_labels = [x_label_mapping[label] for label in ordered_phenotypes]
ax.set_xticklabels(new_labels, fontsize=20)

# 设置坐标轴
plt.xticks(fontsize=20, rotation=0)
plt.yticks(fontsize=20)
plt.ylabel('CytoTRACE', fontsize=20)
plt.xlabel('')
plt.tight_layout()  # 调整布局以避免标签被剪裁

# 保存图片为 PDF
# plt.savefig('project_contacts.pdf', format='pdf')

# 显示图形
plt.show()
```
