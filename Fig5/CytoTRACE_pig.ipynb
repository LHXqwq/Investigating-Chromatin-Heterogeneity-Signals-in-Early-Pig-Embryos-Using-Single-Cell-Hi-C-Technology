```python
#pre-implant_mouse_embryo
#scRNAseq
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statannotations.Annotator import Annotator
from scipy import stats

# 
data = pd.read_csv('pre-implant_mouse_embryo_RPKMCytoTRACE_plot_table.txt', sep='\t')

# 
sns.set(style="whitegrid")
plt.rcParams['font.sans-serif'] = ['DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

trajectory_1 = [
    "early2cell", "mid2cell","late2cell", "4cell","8cell","16cell",   
    "earlyblast","midblast","lateblast"
]

# 12
ordered_phenotypes = trajectory_1

# 
fig, ax = plt.subplots(figsize=(15, 10))  # 

# 
sns.violinplot(x='Phenotype', y='CytoTRACE', data=data,  order=ordered_phenotypes,
               ax=ax, inner="quart", cut=0, scale="width")
sns.stripplot(x='Phenotype', y='CytoTRACE', data=data, order=ordered_phenotypes, 
              color='black', size=5, jitter=True)  # jitter

# 
pairs = [("early2cell", "mid2cell"),("mid2cell", "late2cell"),("late2cell", "4cell"),("4cell", "8cell"),("8cell", "16cell"),
         ("16cell", "earlyblast"),("earlyblast", "midblast"),("midblast", "lateblast")]
annotator = Annotator(ax, pairs, data=data, x="Phenotype", y="CytoTRACE", order=ordered_phenotypes)

# 
chosen_test = 'Mann-Whitney'  # 
annotator.configure(test=chosen_test, text_format='star', line_height=0.01, 
                    line_width=1, loc='outside', fontsize=20)
annotator.apply_and_annotate()

#  x 
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylabel('CytoTRACE', fontsize=20)
plt.xlabel('')
plt.tight_layout()  # 

#  PDF
# plt.savefig('project_contacts.pdf', format='pdf')

# 
plt.show()
```

```python
#scRNAseq
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statannotations.Annotator import Annotator
from scipy import stats

# 
data = pd.read_csv('PA_IVF_198_CytoTRACE_results.txt', sep='\t')

# 
sns.set(style="whitegrid")
plt.rcParams['font.sans-serif'] = ['DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# 
color_mapping = {
    "PA-MII": "#ffe119",
    "PA-2C": "#3cb44b",
    "PA-4C": "#e6194b",
    "IVF-Zygote": "#ffe119",
    "IVF-2C": "#3cb44b",
    "IVF-4C": "#e6194b"
}

# 
# 1
trajectory_1 = [
    "PA-MII", "PA-2C", "PA-4C"
]

# 2
trajectory_2 = [
     "IVF-Zygote","IVF-2C", "IVF-4C"
]

# 12
ordered_phenotypes = trajectory_1 + trajectory_2

# 
fig, ax = plt.subplots(figsize=(15, 6))  # 

# 
sns.violinplot(x='Phenotype', y='CytoTRACE', data=data, 
               palette=color_mapping, order=ordered_phenotypes,
               ax=ax, inner="quart", cut=0, scale="width")
sns.stripplot(x='Phenotype', y='CytoTRACE', data=data, order=ordered_phenotypes, color='black', size=5)

# 
pairs = [("PA-MII", "PA-2C"),("PA-2C", "PA-4C"),("IVF-Zygote", "IVF-2C"),("IVF-2C", "IVF-4C")]
annotator = Annotator(ax, pairs, data=data, x="Phenotype", y="CytoTRACE", order=ordered_phenotypes)

chosen_test='t-test_welch'
if chosen_test == 't-test_ind':
    annotator.configure(test='t-test_ind', text_format='star', line_height=0.01, line_width=1, loc='outside',fontsize=20)
    print(" t- ( t )")
elif chosen_test == 't-test_welch':
    annotator.configure(test='t-test_welch', text_format='star', line_height=0.01, line_width=1, loc='outside',fontsize=20)
    print(" Welch's t-")
else:
    annotator.configure(test='Mann-Whitney', text_format='star', line_height=0.01, line_width=1, loc='outside',fontsize=20)
    print(" Mann-Whitney U ")

annotator.apply_and_annotate()

#  x 
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylabel('CytoTRACE',fontsize=20)
plt.xlabel('')
plt.tight_layout()  # 

#  PDF
# plt.savefig('project_contacts.pdf', format='pdf')

# 
plt.show()
```
