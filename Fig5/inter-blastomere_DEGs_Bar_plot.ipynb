```python
import matplotlib.pyplot as plt
from IPython.display import SVG, display

plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['svg.fonttype'] = 'none'
plt.rcParams['svg.hashsalt'] = 'hello'

# 数据和标签
data = [618, 1181, 697, 2116]
labels = ['PA-2C', 'PA-4C', 'IVF-2C', 'IVF-4C']

# 绘制柱状图
plt.figure(figsize=(8, 6))
plt.bar(labels, data, color=['#3cb44b', '#e6194b', '#3cb44b', '#e6194b'])

# 添加标题和标签
plt.title("Inter-blastomere DEGs",fontsize=20)
plt.xlabel("")
plt.ylabel("Counts",fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylim(0, 2250)
# 显示数据值
for i, value in enumerate(data):
    plt.text(i, value + 50, str(value), ha='center', fontsize=15)

# 显示图表
plt.tight_layout()
plt.savefig('Inter-blastomere_DEGs.svg', dpi=300, bbox_inches='tight', transparent=True)

```
